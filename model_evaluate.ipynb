# Evaluate the Model

```python
import pandas as pd
import mlflow
import joblib
from sklearn.metrics import mean_absolute_error,root_mean_squared_error, r2_score
```

```python
x_evaluate = pd.read_csv(r"N:\GUVI\Smart_Premium\Dataset\test.csv", index_col="id")
y_evaluate = pd.read_csv(r"N:\GUVI\Smart_Premium\Dataset\sample_submission.csv", index_col="id")
```

```python
from scipy import stats

# Box-Cox transform Annual Income
def boxcox_transform(X, l = 0.5):
    X = X.copy()
    X[:, 1] = stats.boxcox(X[:, 1], lmbda=l)
    return X
```

```python
pre_pipeline = joblib.load('preprocessing_pipeline.pkl')
x_evaluate_prepared = pre_pipeline.transform(x_evaluate)
x_evaluate_prepared
```

```python
y_evaluate
```

```python
mlflow.set_tracking_uri("https://dagshub.com/nithis127/Smart_Premium.mlflow")
```

```python
rfr_model = mlflow.sklearn.load_model(model_uri= 'models:/sp_rfr/Production')

y_pred = rfr_model.predict(x_evaluate_prepared)
print("MAE:", mean_absolute_error(y_evaluate,y_pred))
print("RMSE:", root_mean_squared_error(y_evaluate,y_pred))
print("R2 :", r2_score(y_evaluate,y_pred))
```
